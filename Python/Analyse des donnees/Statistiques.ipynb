```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Lecture du dataset (Remplace par ton fichier)
df = pd.read_csv("data fil rouge.csv")
df

```

1. Corrélation de Pearson
Question

L'âge influence-t-il le profit net ?




Hypothèses
H0 : pas de corrélation entre Age et Profit_Net                                              
H1 : il existe une corrélation entre Age et Profit_Net.

```python
from scipy.stats import pearsonr

corr, p_value = pearsonr(df["Age"], df["Profit_Net"])

print("Corrélation :", corr)
print("P-value :", p_value)

if p_value < 0.05:
    print("Rejet de H0")
else:
    print("On ne rejette pas H0")
```

2. Test du Khi-deux (Chi-Square)
Question

La note Google influence-t-elle le niveau des ventes ?

Préparation

Transformer les ventes en catégories.

```python
df["Vente_Categorie"] = pd.qcut(
    df["Ventes_Brutes"],
    q=3,
    labels=["Faible", "Moyenne", "Elevée"]
)
```

Hypothèses
H0 : Note Google et niveau de ventes sont indépendants                                                        
H1 : elles sont liées

```python
from scipy.stats import chi2_contingency

table = pd.crosstab(
    round(df["Note_Google_Maps"]),
    df["Vente_Categorie"]
)

chi2, p, dof, expected = chi2_contingency(table)

print("Chi2 :", chi2)
print("P-value :", p)
```

3. Test t de Student
Question

Les clients avec une note Google élevée génèrent-ils plus de profit ?

Préparation

```python
groupe1 = df[df["Note_Google_Maps"] >= 4]["Profit_Net"]

groupe2 = df[df["Note_Google_Maps"] < 4]["Profit_Net"]
```

Hypothèses
H0 : les profits moyens sont identiques                                        
H1 : les profits moyens sont différents

```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(
    groupe1,
    groupe2,
    equal_var=False
)

print("T-stat :", t_stat)
print("P-value :", p_value)
```
