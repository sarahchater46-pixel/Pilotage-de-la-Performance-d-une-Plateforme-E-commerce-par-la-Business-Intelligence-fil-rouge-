```python

import numpy as np
import random
from datetime import datetime, timedelta

# 1. Configuration des paramètres marocains
n_rows = 80000

regions_villes = {
    "Casablanca-Settat": ["Casablanca", "Mohammedia", "Berrechid", "El Jadida", "Settat"],
    "Rabat-Salé-Kénitra": ["Rabat", "Salé", "Kénitra", "Skhirat", "Témara"],
    "Marrakech-Safi": ["Marrakech", "Safi", "Essaouira"],
    "Fès-Meknès": ["Fès", "Meknès", "Taza", "Ifrane"],
    "Tanger-Tétouan-Al Hoceima": ["Tanger", "Tétouan", "Al Hoceima", "Larache"],
    "Souss-Massa": ["Agadir", "Inezgane", "Tiznit"],
    "Béni Mellal-Khénifra": ["Béni Mellal", "Khénifra", "Kasba Tadla"]
}

categories = {
    "Electronics": ["Phones", "Accessories", "Printers", "Electronic Games", "Laptops"],
    "Clothing": ["Saree", "Hankerchief", "Kurti", "Trousers", "Skirt", "T-shirt", "Leggings"],
    "Furniture": ["Chairs", "Tables", "Bookcases", "Furnishings"]
}

payment_modes = ["Paiement à la livraison", "Carte bancaire", "Virement bancaire", "Paiement en plusieurs fois"]
names = ["Youssef", "Amina", "Mohamed", "Fatima", "Omar", "Khadija", "Ahmed", "Sara", "Rachid", "Nadia", "Mustapha", "Zineb", "Karim", "Salma", "Driss", "Meriem"]

# 2. Génération des données
data = []

for i in range(n_rows):
    # ID
    order_id = f"M-{10000 + i}"
    
    # Dates (sur les 2 dernières années)
    start_date = datetime(2022, 1, 1)
    date = start_date + timedelta(days=random.randint(0, 800))
    
    # Client et Localisation
    name = random.choice(names)
    region = random.choice(list(regions_villes.keys()))
    city = random.choice(regions_villes[region])
    
    # Produit
    cat = random.choice(list(categories.keys()))
    sub_cat = random.choice(categories[cat])
    
    # Logique de prix et quantité
    qty = random.randint(1, 15)
    if cat == "Electronics":
        sales = random.randint(500, 15000)
    elif cat == "Furniture":
        sales = random.randint(300, 8000)
    else: # Clothing
        sales = random.randint(20, 1200)
        
    # Calcul du profit (avec risque de perte sur paiement à la livraison)
    pay_mode = random.choice(payment_modes)
    margin = random.uniform(-0.2, 0.4) # Entre 20% de perte et 40% de gain
    
    # Simulation d'un problème réel au Maroc : Le Cash on Delivery augmente les retours/pertes
    if pay_mode == "Paiement à la livraison" and random.random() < 0.15:
        profit = - (sales * 0.1) # Perte liée au transport/retour
    else:
        profit = int(sales * margin)

    data.append([order_id, date, name, region, city, sales, profit, qty, cat, sub_cat, pay_mode])

# 3. Création du DataFrame et export
df_final = pd.DataFrame(data, columns=[
    "Order ID", "Order Date", "CustomerName", "State", "City", 
    "Total Sales", "Profit", "Quantity", "Category", "Sub-Category", "PaymentMode"
])

# Sauvegarde
df_final.to_csv("maroc_ecommerce_80k.csv", index=False)
print("Dataset de 80 000 lignes généré avec succès !")
```

```python
import pandas as pd

# On part de la table de 80 000 lignes que vous avez générée (df_final)
# Si vous rechargez le fichier : df_final = pd.read_csv("maroc_ecommerce_80k.csv")

# 1. TABLE DIM_GEOGRAPHY (Localisation)
dim_geo = df_final[['State', 'City']].drop_duplicates().reset_index(drop=True)
dim_geo['Location_ID'] = dim_geo.index + 1
dim_geo = dim_geo[['Location_ID', 'State', 'City']]

# 2. TABLE DIM_PRODUCT (Produits)
dim_product = df_final[['Category', 'Sub-Category']].drop_duplicates().reset_index(drop=True)
dim_product['Product_ID'] = dim_product.index + 1
dim_product = dim_product[['Product_ID', 'Category', 'Sub-Category']]

# 3. TABLE DIM_CUSTOMER (Clients)
# On simule des IDs clients pour les noms uniques
dim_customer = pd.DataFrame({'CustomerName': df_final['CustomerName'].unique()})
dim_customer['Customer_ID'] = dim_customer.index + 1
dim_customer = dim_customer[['Customer_ID', 'CustomerName']]

# 4. TABLE FACT_SALES (La table de faits)
# On remplace les noms par les IDs pour créer les relations
fact_sales = df_final.merge(dim_geo, on=['State', 'City']) \
                     .merge(dim_product, on=['Category', 'Sub-Category']) \
                     .merge(dim_customer, on='CustomerName')

# On ne garde que les IDs et les métriques
fact_sales = fact_sales[['Order ID', 'Order Date', 'Customer_ID', 'Location_ID', 
                         'Product_ID', 'Total Sales', 'Profit', 'Quantity', 'PaymentMode']]

# EXPORTATION DES 4 FICHIERS
dim_geo.to_csv("Dim_Geography.csv", index=False)
dim_product.to_csv("Dim_Product.csv", index=False)
dim_customer.to_csv("Dim_Customer.csv", index=False)
fact_sales.to_csv("Fact_Sales.csv", index=False)

print("Modélisation terminée : 4 tables prêtes pour Power BI ou SQL !")
```

```python
# Exportation au format CSV (le plus standard pour la Data)
fact_sales.to_csv("Fact_Sales.csv", index=False, encoding='utf-8-sig')
dim_geo.to_csv("Dim_Geography.csv", index=False, encoding='utf-8-sig')
dim_product.to_csv("Dim_Product.csv", index=False, encoding='utf-8-sig')
dim_customer.to_csv("Dim_Customer.csv", index=False, encoding='utf-8-sig')

print("Fichiers CSV exportés avec succès !")
```
